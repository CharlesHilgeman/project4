```python
import pandas as pd
```

```python
train_data = pd.read_csv("adult_data.csv")
test_data = pd.read_csv("adult_test.csv")
```

```python
test_data
```

```python
# FIXING COLUMN NAMES

train_data = train_data.rename(columns={" education-num":"education_num"," marital-status":"marital_status"," capital-gain":"capital_gain"," capital-loss":"capital_loss"," hours-per-week":"hours"," native-country":"native_country"})
train_data = train_data.rename(columns={" workclass":"workclass"," fnlwgt":"fnlwgt"," education":"education"," occupation":"occupation"," relationship":"relationship"," race":"race"," sex":"sex"," salary":"salary"})

test_data = test_data.rename(columns={" education-num":"education_num"," marital-status":"marital_status"," capital-gain":"capital_gain"," capital-loss":"capital_loss"," hours-per-week":"hours"," native-country":"native_country"," workclass":"workclass"," fnlwgt":"fnlwgt"," education":"education"," occupation":"occupation"," relationship":"relationship"," race":"race"," sex":"sex"," salary":"salary"})
```

```python
train_data["occupation"].unique()
```

```python
X_train = pd.get_dummies(train_data.drop(columns=['salary']))
X_train
```

```python
y_train = train_data["salary"]
y_train
```

```python
X_test = pd.get_dummies(test_data.drop(columns=['salary']))
X_test
```

```python
y_test = test_data["salary"]
y_test
```

```python
res = [x for x in X_train + X_test if x not in X_train or x not in X_test]
 
print(res)
```

```python
X_test["native_country_ Holand-Netherlands"] = 0
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()

classifier.fit(X_train, y_train)

print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)

print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
classifier = LogisticRegression()

classifier.fit(X_train_scaled, y_train)

print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)

print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```
