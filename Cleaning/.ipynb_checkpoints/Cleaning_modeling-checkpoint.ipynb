```python
import pandas as pd
```

```python
train_data = pd.read_csv("adult_data.csv")
```

```python
train_data
```

```python
# FIXING COLUMN NAMES

train_data = train_data.rename(columns={" education-num":"education_num"," marital-status":"marital_status"," capital-gain":"capital_gain"," capital-loss":"capital_loss"," hours-per-week":"hours"," native-country":"native_country"})
train_data = train_data.rename(columns={" workclass":"workclass"," fnlwgt":"fnlwgt"," education":"education"," occupation":"occupation"," relationship":"relationship"," race":"race"," sex":"sex"," salary":"salary"})
```

```python
train_data["occupation"].unique()
```

```python
# CLEANING DATAFRAME

train_data = train_data[train_data["workclass"]!= " ?"]
train_data = train_data[train_data["occupation"]!= " ?"]
train_data = train_data.replace(' ', '', regex=True)
train_data.reset_index()
train_data
```

```python
train_data = train_data.drop(columns={"fnlwgt","education_num","relationship","capital_gain","capital_loss"})
train_data
```

```python
# CREATE X TRAIN

X_train = pd.get_dummies(train_data.drop(columns=['salary']))
X_train
```

```python
# CREATE Y TRAIN

y_train = train_data["salary"]
y_train
```

```python
# SPLIT DATA

from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(X_train, y_train, random_state=1)
```

```python
# LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression

classifier1 = LogisticRegression()

classifier1.fit(X_train1, y_train1)

print(f"Training Data Score: {classifier1.score(X_train1, y_train1)}")
print(f"Testing Data Score: {classifier1.score(X_test1, y_test1)}")
```

```python
# RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier

clf1 = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train1, y_train1)

print(f'Training Score: {clf1.score(X_train1, y_train1)}')
print(f'Testing Score: {clf1.score(X_test1, y_test1)}')

# clf1.predict(new stuff) == youre salary will be >50k
```

```python
# SCALE DATA

from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler().fit(X_train1)
X_train_scaled1 = scaler1.transform(X_train1)

X_test_scaled1 = scaler1.transform(X_test1)
```

```python
# LOGISTIC REGRESSION WITH SCALED DATA

classifier2 = LogisticRegression()

classifier2.fit(X_train_scaled1, y_train1)

print(f"Training Data Score: {classifier2.score(X_train_scaled1, y_train1)}")
print(f"Testing Data Score: {classifier2.score(X_test_scaled1, y_test1)}")
```

```python
# RANDOM FOREST WITH SCALED DATA

clf2 = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train1, y_train1)

print(f'Training Score: {clf2.score(X_train_scaled1, y_train1)}')
print(f'Testing Score: {clf2.score(X_test_scaled1, y_test1)}')
```
